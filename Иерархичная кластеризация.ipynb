```python
import numpy as np

Z = np.array(
    [
        [85,3],
        [24,55],
        [70,15],
        [15,78],
        [60,12],
        [80,30],
        [5,70],
        [10,80],
        [70,10],
        [30,91],
    ]
)
```

```python
Z
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

labels = range(1, 11)
plt.figure(figsize=(10, 7))
plt.scatter(Z[:,0],
            Z[:,1], 
            label='Истинная позиция')

for label, x, y in zip(labels, Z[:, 0], Z[:, 1]):
    plt.annotate(
        label,
        xy=(x, y),
        xytext=(-3, 3),
        textcoords='offset points',
        ha='left',
        va='bottom')
plt.show()
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
from matplotlib import pyplot as plt

linked = linkage(Z, 'single')

labelList = range(1, 11)

plt.figure(figsize=(10, 7))
dendrogram(linked,
           orientation='top',
           labels=labelList,
           distance_sort='descending',
           show_leaf_counts=True)
plt.show()
```
